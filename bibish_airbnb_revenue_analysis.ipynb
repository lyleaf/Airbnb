```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('airbnb.csv')

df = df[df['Type'] == 'Reservation']
# Plot the data
df.head()
```

```python
df.loc[df['Cleaning Fee'] >= 150, 'Cleaning Fee'] = 210
```

```python
df.loc[df['Cleaning Fee'] == 100, 'Cleaning Fee'] = 150
```

```python
df['Cleaning Fee'].values
```

```python
df = df[df['Cleaning Fee'] > 100]
```

```python
df['Profit'] = df['Amount'] - df['Cleaning Fee']
```

```python
df.head()
```

```python
df.head(100)
```

```python
# Plot the data
plt.plot(df['Date'], df['Profit'])
plt.show()
```

```python
df['date'] = pd.to_datetime(df['Date'])
```

```python
df_monthly_profit = df.groupby(df['date'].dt.month)['Profit'].sum()
```

```python
df_monthly_cost = df.groupby(df['date'].dt.month)['Cleaning Fee'].sum()
```

```python
type(df_monthly_profit)
```

```python
# Plot the data
plt.bar(df_monthly_profit.index, df_monthly_profit.values, color='blue', width=0.5)

plt.show()
```

```python
# Plot the data
plt.bar(df_monthly_profit.index, df_monthly_cost.values, color='red', width=0.5)

plt.show()
```
